```python
import pandas as pd
```

```python
df = pd.read_csv('../data/input/IMDB-Movie-Data.csv')
df.info()
```

```python
def movie_votes(votes):
    if 0 < votes <= 999:
        return 1
    elif 999 < votes <= 9999:
        return 2
    elif 9999 < votes <= 99999:
        return 3
    elif 99999 < votes <= 999999:
        return 4
    elif 999999 < votes:
        return 5
```

```python
df['bins'] = df.apply(lambda x : movie_votes( x['Votes']), axis=1)
df
```

```python
df['Rev_Mins'] = df['Revenue (Millions)']/df['Runtime (Minutes)']
df
```

```python
#Challenge 3
#We want to create a new ranking where we add 1 point if the genre is thriller 
#but subtract 1 point if the genre is comedy.

def new_rating(Genre,Rating):
    if 'Thriller' in Genre:
        return (Rating+1)
    elif 'Comedy' in Genre:
        return (Rating-1)
    else:
        return Rating
```

```python
df['new_rating'] = df.apply(lambda x: new_rating(x['Genre'],x['Rating']),axis=1)
df
```

```python
#Challenge 4
#We want to know if the sum of the ASCII value of every character of the movie title 
#divided by the number of votes retrieve a prime number
#...remember, prime numbers are integers.

```

```python
#Primero averiguo la longitud del titulo
df['title_len'] = df['Title'].str.len()
df
```

```python
#después lo dividimos entre el número de votos
df['Title_votes'] = df['title_len']/df['Votes']
df
```

```python
#Revisamos si hay números primos
def isprime(n):
    n = abs(int(n))

    if n < 2:
        return False

    if n == 2: 
        return True    
    
    if not n & 1: 
        return False

    for x in range(3, int(n**0.5) + 1, 2):
        if n % x == 0:
            return False

    return True

df['Prime_Number'] = df['Title_votes'].apply(isprime)
df
```

```python
#Por último revisamos si hay número primos pero todos devuelven falso
df['Prime_Number'].unique()
```
